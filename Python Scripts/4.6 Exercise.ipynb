# Merge Dataframe

### <b>Rationale:</b> The column "product_id" exists in both the "orders_products_combined" and the cleaned products dataset; therefore, it is a reliable common key for merging. Since our objective is to enrich each order record with corresponding product details, and only records with matching product_id values in both datasets are meaningful for analysis, an inner merge is appropriate. This ensures the final dataframe retains only valid, complete records where product metadata is available.

```python
# Test merge without creating a real dataframe
pd.merge(orders_products_combined, df_prods, on= ['product_id'], how='inner', indicator= 'match')
```

```python
# Real Merge
ords_prods_merge = orders_products_combined.merge(df_prods, on = ['product_id'], indicator = 'match')
```

```python
# drop useless '_merge' column
ords_prods_merge.drop('_merge', axis = 1)
```
