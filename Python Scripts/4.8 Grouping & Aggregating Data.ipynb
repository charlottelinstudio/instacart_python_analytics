### Step 4. Create a loyalty flag for existing customers

```python
# Create a new 'max_order' column, split the data into 'user_id' column, and apply transform() function to generate the maximum orders for each user.
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check results
ords_prods_merge[['user_id', 'order_number', 'max_order']].head(25)
```

#### The above list shows all User IDs of 1, which may be due to the limited sample size. Therefore, I will run another code to look at a more random sample from the entire dataframe.

```python
ords_prods_merge[['user_id', 'order_number', 'max_order']].sample(25)
```

#### Apply filters to flag

```python
# 'Loyal customer' > 40 orders; 'Regular customer' = 11-40 orders; 'New customer' < 11 orders.
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

#### Check frequency

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

#### Check output
- I'm using sample() again to broaden the list

```python
ords_prods_merge[['user_id', 'max_order', 'loyalty_flag']].sample(25)
```

## Step 6. Target diffrent types of spenders

```python
# Create a spending flag with a new 'avg_spend_user' column
ords_prods_merge['avg_spend_user'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check results
ords_prods_merge[['user_id', 'prices', 'avg_spend_user']].sample(25)
```

```python
# Apply filter to the new 'avg_spend_user' column: 'Low spender' < 10; 'High spender' >= 10.
ords_prods_merge.loc[ords_prods_merge['avg_spend_user'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_spend_user'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequency
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check output
ords_prods_merge[['user_id', 'avg_spend_user', 'spending_flag']].sample(25)
```

## Step 7. Send relevant notifications to users by order frequency

```python
# Create an order frequency flag that marks the regularity of a users ordering behavior according to the median in the days_since_prior_order column.
ords_prods_merge['ord_regularity_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check results 
ords_prods_merge[['user_id', 'days_since_prior_order', 'ord_regularity_median']].sample(25)
```

```python
# Apply filters to 'ord_regularity_median" column: 'Non-frequent customer' > 20; 'Regular customer' = 10-20; 'Frequent customer' <= 10.
ords_prods_merge.loc[ords_prods_merge['ord_regularity_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['ord_regularity_median'] > 10) & (ords_prods_merge['ord_regularity_median'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['ord_regularity_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check results
ords_prods_merge[['user_id', 'days_since_prior_order', 'ord_regularity_median', 'frequency_flag']].sample(25)
```

```python
# Check frequency
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
